### (1) 데이터 분할1 : x, y

```python
x = data.drop("Activity_dynamic", axis = 1)
y = data.loc[:, "Activity_dynamic"]
```

### (1) 알고리즘1 : AutoML

```python
!pip install mljar-supervised
```

```python
from supervised.automl import AutoML

automl = AutoML(mode = "Compete")
automl.fit(x, y)
```

```python
test = pd.read_csv("/content/drive/MyDrive/딥러닝/5차 미프/keggle/test_dataset.csv")
test.head()
```

```python
sample = pd.read_csv("/content/drive/MyDrive/딥러닝/5차 미프/keggle/sample_submission.csv")
predictions = automl.predict_all(test)
predictions
```

```python
sample["Activity"] = predictions["label"]
sample
```
